```python
def clean_data(titanic):
    new_feature = pd.DataFrame()
    new_feature['PassengerId'] = titanic['PassengerId']
    titanic["Age"] = titanic['Age'].fillna(titanic['Age'].median())
    #child 
    new_feature['child'] = titanic['Age'].apply(lambda x: 1 if x < 15 else 0)
    #sex
    new_feature['sex'] = titanic['Sex'].apply(lambda x: 1 if x=='male' else 0)
    #embark
    def getEmbark(embark):
        return pd.get_dummies(embark, prefix='Embarked')
    new_feature = new_feature.join(getEmbark(titanic['Embarked']))
    # familysize
    new_feature['family_size'] = titanic['SibSp'] + titanic['Parch'] + 1
    # cabin
    titanic['Cabin'] = titanic['Cabin'].fillna('U')
    def getCabin(cabin):
        return pd.get_dummies(cabin, prefix = 'Cabin')
    new_feature = new_feature.join(getCabin(titanic['Cabin']))  
    #pclass
    def getClass(Pclass):
        return pd.get_dummies(Pclass, prefix = 'Class')
    new_feature = new_feature.join(getClass(titanic['Pclass']))
    
    #name
    def getTitle(name):
        title = pd.DataFrame()
        title['Title'] = name.map(lambda name: name.split(',')[1].split('.')[0].strip())
        Title_Dict={
              "Capt": "Officer",
              "Col": "Officer",
              "Major": "Officer",
              "Jonkheer": "Royalty",
              "Don": "Royalty",
               "Sir": "Royalty",
               "Dr": "Officer",
               "Rev": "Officer",
               "the Countess": "Royalty",
               "Dona": "Royalty",
               "Mme": "Mrs",
               "Mlle": "Miss",
               "Ms": "Mrs",
               "Mr": "Mr",
               "Mrs": "Mrs",
               "Miss": "Miss",
                "Master": "Master",
                "Lady": "Royalty"}
        title['Title'] = title['Title'].map(Title_Dict)
        return pd.get_dummies(title['Title'])
    
    new_feature = new_feature.join(getTitle(titanic['Name']))
   
    
    #Fare 
    new_feature['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())
    return new_feature  
```
